```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import time
import warnings
warnings.filterwarnings("ignore")

```

```python
df_countries = pd.read_csv('countries.csv')
#df_countries.head()
```

```python
df_names = df_countries[['Country']]
#df_names.to_csv('country_names.csv', index=False)
#print(df_names.head())
```

```python
df_name_copy = df_names.copy()
# make everything lowercase
df_name_copy['Country'] = df_name_copy['Country'].str.lower()
print(df_name_copy.head())

# start producing strings of increasing length by going through everything starting from A. imclude spaces
def generate_strings(length):
    if length == 1:
        return [chr(i) for i in range(97, 123)]  # a-z
    else:
        prev_strings = generate_strings(length - 1)
        new_strings = []
        for s in prev_strings:
            for i in range(97, 123):
                new_strings.append(s + chr(i))
            new_strings.append(s + ' ')  # add space
        return new_strings
length = 1
start_time = time.time()
while df_name_copy.shape[0] >0:
    # generate strings of increasing length
    strings = generate_strings(length)
    print(f"Length: {length}, Strings: {len(strings)}, exponential: {27 ** length-27**(length-1)}")
    print(strings)
    print(df_name_copy.head())
    # check if the any strings are in the country names. if they are, remove them from the dataframe
    df_name_copy = df_name_copy[~df_name_copy['Country'].isin(strings)]
    print(df_name_copy.head())
    # print the number of rows in the dataframe
    print(f"Length: {length}, Rows: {df_name_copy.shape[0]}")
    # increment length
    length += 1
    if length > 3:
        break
# print the time taken
end_time = time.time()
print(f"Time taken: {end_time - start_time} seconds")
```
